```python
# Convert 'Dt_Customer' to datetime objects
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)

# Create 'Age' column from 'Year_Birth' (Assuming current year is 2026)
df['Age'] = 2026 - df['Year_Birth']

# Create 'Total_Spent' by summing all spending columns
spending_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Spent'] = df[spending_cols].sum(axis=1)

# Create 'Total_Children' and 'Is_Parent'
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
df['Is_Parent'] = np.where(df['Total_Children'] > 0, 1, 0)

# Check the results
display(df[['Age', 'Total_Spent', 'Total_Children', 'Is_Parent']].head())
```

```python
#1. Customer Information Table
# Adding the new columns: age, total_children, and is_parent here
customer_info = df[['ID', 'Year_Birth', 'Education', 'Marital_Status', 'Age', 'Total_Children', 'Is_Parent']].copy()
customer_info.columns = ['id', 'birth_year', 'education', 'marital_status', 'age', 'total_children', 'is_parent']

# 2. Financial Details Table
financial_details = df[['ID', 'Income']].copy()
financial_details.columns = ['id', 'income']

# 3. Purchase Behavior Table
# Adding the new column: total_spent here
purchase_behavior = df[['ID', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 
                        'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'Total_Spent']].copy()
purchase_behavior.columns = ['id', 'recency', 'mnt_wines', 'mnt_fruits', 'mnt_meat', 
                             'mnt_fish', 'mnt_sweet', 'mnt_gold', 'total_spent']

# 4. Purchase Channels Table
purchase_channels = df[['ID', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 
                        'NumStorePurchases', 'NumWebVisitsMonth']].copy()
purchase_channels.columns = ['id', 'deals_purchases', 'web_purchases', 'catalog_purchases', 
                             'store_purchases', 'web_visits_month']

# 5. Campaign Response Table
campaign_response = df[['ID', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 
                        'AcceptedCmp4', 'AcceptedCmp5', 'Response']].copy()
campaign_response.columns = ['id', 'cmp1', 'cmp2', 'cmp3', 'cmp4', 'cmp5', 'response']

# 6. Customer Engagement Table
customer_engagement = df[['ID', 'Complain', 'Dt_Customer']].copy()
customer_engagement.columns = ['id', 'complain', 'enrollment_date']

print(" All tables created and renamed with ID as the primary key.")
# To check one of the tables:
customer_info.head()
```

```python
# Re-creating merged_all properly
merged_all = pd.merge(customer_info, financial_details, on='id')
merged_all = pd.merge(merged_all, purchase_behavior, on='id')

# Define columns and calculate correlation
# Making sure all columns exist in the new merged_all
cols_to_check = ['income', 'age', 'total_spent', 'total_children', 'mnt_wines', 'mnt_meat', 'mnt_gold']
correlation_matrix = merged_all[cols_to_check].corr()

# Create mask to show only lower triangle (makes it cleaner)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Create the Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='RdBu_r', 
            fmt='.2f', square=True, linewidths=0.5)

plt.title('Unique Feature Correlations', fontweight='bold', size=15)
plt.tight_layout()
plt.show()

# Print strong correlations (The "Insights" part)
print("\n--- Strong Correlations Analysis ---")
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if abs(corr_val) > 0.5:
            print(f"{correlation_matrix.columns[i]} & {correlation_matrix.columns[j]}: {corr_val:.3f}")
```
